```python
# Jupyter Notebook code
import pandas as pd
import matplotlib.pyplot as plt
import os

# --- Step 1: Load Dataset ---
dataset_path = os.path.join("..", "data", "VIIRS-nighttime-lights-2013m1to2024m5-level2.csv")

# Use latin1 encoding to avoid Unicode errors
df = pd.read_csv(dataset_path, encoding='latin1')

# Quick check
print("Columns:", df.columns.tolist())
print(df.head())

# --- Step 2: Filter for Asia ---
asia_countries = [
    "AFG", "ARM", "AZE", "BHR", "BGD", "BTN", "BRN", "KHM", "CHN", "CYP",
    "GEO", "IND", "IDN", "IRN", "IRQ", "ISR", "JPN", "JOR", "KAZ", "KWT",
    "KGZ", "LAO", "LBN", "MYS", "MDV", "MNG", "MMR", "NPL", "PRK", "OMN",
    "PAK", "PSE", "PHL", "QAT", "SAU", "SGP", "KOR", "LKA", "SYR", "TWN",
    "TJK", "THA", "TLS", "TUR", "TKM", "ARE", "UZB", "VNM", "YEM"
]

df_asia = df[df['iso'].isin(asia_countries)]

# --- Step 3: Aggregate nlsum by year ---
df_yearly = df_asia.groupby('year')['nlsum'].sum().reset_index()

# --- Step 4: Plot ---
plt.figure(figsize=(10,6))
plt.plot(df_yearly['year'], df_yearly['nlsum'], marker='o', color='orange')
plt.title('Total Nightlight in Asia (2013-2024)')
plt.xlabel('Year')
plt.ylabel('Nightlight Sum')
plt.grid(True)
plt.tight_layout()

# Optional: save plot
plot_path = os.path.join("..", "data", "asia_nightlight_plot.png")
plt.savefig(plot_path)

plt.show()
```

```python
df_asia.to_csv("../data/asia_nightlight_data.csv", index=False)
```

```python
df_monthly = df_asia.groupby(['year','month'])['nlsum'].sum().reset_index()

# Example plot
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
for year in df_monthly['year'].unique():
    monthly_data = df_monthly[df_monthly['year']==year]
    plt.plot(monthly_data['month'], monthly_data['nlsum'], marker='o', label=year)
plt.title("Monthly Nightlight Sum in Asia (2013-2024)")
plt.xlabel("Month")
plt.ylabel("Nightlight Sum")
plt.legend()
plt.show()
```

```python
df_country = df_asia.groupby('iso')['nlsum'].sum().reset_index().sort_values(by='nlsum', ascending=False)
plt.figure(figsize=(12,6))
plt.bar(df_country['iso'], df_country['nlsum'], color='skyblue')
plt.title("Total Nightlight per Asian Country (2013-2024)")
plt.xlabel("Country ISO Code")
plt.ylabel("Nightlight Sum")
plt.xticks(rotation=90)
plt.show()
```
